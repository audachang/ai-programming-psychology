## 2. Numpy vs Lists
Performance comparison to see why we use Numpy for data.

````python
import time

size = 1000000

# List
t0 = time.time()
l = [i for i in range(size)]
l_sq = [i**2 for i in l]
print(f"List time: {time.time()-t0:.5f}s")

# Numpy
t0 = time.time()
arr = np.arange(size)
arr_sq = arr**2
print(f"Numpy time: {time.time()-t0:.5f}s")
````
